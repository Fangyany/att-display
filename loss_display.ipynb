```python
import numpy as np
import torch
```

```python
LSTM = open('LSTM/LSTM/log', 'r')

LSTMATT = open('LSTMATT/LSTMATT/log', 'r')

LSTMGAN = open('LSTMGAN/LSTMGAN/log', 'r')

LSTMGAT = open('LSTMGAT/LSTMGAT/log', 'r')

CNN = open('CNN/CNN/log', 'r')

CNNGAN = open('CNNGAN/CNNGAN/log', 'r')
```

```python
m = 7

loss_LSTM = []
for line in LSTM:
    data = line.split()
    if 'ade1' in data:
        loss_LSTM.append(float(data[m][0:6]))
  

loss_LSTMATT = []
for line in LSTMATT:
    data = line.split()
    if 'ade1' in data:
        loss_LSTMATT.append(float(data[m][0:6]))    
        
loss_LSTMGAN = []
for line in LSTMGAN:
    data = line.split()
    if 'ade1' in data:
        loss_LSTMGAN.append(float(data[m][0:6]))    
        
loss_LSTMGAT = []
for line in LSTMGAT:
    data = line.split()
    if 'ade1' in data:
        loss_LSTMGAT.append(float(data[m][0:6]))
        
loss_CNN = []
for line in CNN:
    data = line.split()
    if 'ade1' in data:
        loss_CNN.append(float(data[m][0:6]))
        
loss_CNNGAN = []
for line in CNNGAN:
    data = line.split()
    if 'ade1' in data:
        loss_CNNGAN.append(float(data[m][0:6]))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


sns.set()

plt.plot(loss_LSTM, '--',label='LSTM', color='r')
plt.plot(loss_LSTMATT, label='LSTMATT', color='y')
plt.plot(loss_LSTMGAT, label='LSTMGAT', color='y')
plt.plot(loss_LSTMGAN, label='LSTM+GAN', color='r')
plt.plot(loss_CNN, '--', label='CNN', color='g')
plt.plot(loss_CNNGAN, label='CNN+GAN', color='g')

# plt.ylabel('average displacement error')
plt.ylabel('final displacement error')

plt.xlabel('iteration number')
plt.legend()
plt.rcParams['figure.figsize'] = (6,6)
```
